```python
from keras.models import load_model
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
model = load_model('HSVmodel.h5')
```

```python
dataset_path = r"C:\Users\91909\Desktop\ML\DATA\Plant_Disease_New\color"
img_path = r"C:\Users\91909\Desktop\ML\DATA\Plant_Disease_New\color\Tomato___Late_blight\1ba00fc3-a45b-4d23-94da-5e090ba7f3d3___RS_Late.B 6403.JPG"
```

```python
image = cv2.imread(img_path)
image = cv2.resize(image, (224, 224))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = image / 255.0
plt.imshow(image)
```

```python
if image.dtype != np.uint8:
    image = (image * 255).astype(np.uint8)

image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(image)
```

```python
input_image = np.expand_dims(image, axis=0)
```

```python
classes = sorted(os.listdir(dataset_path))
classes
```

```python
pred = model.predict(input_image)
```

```python
pred_name = classes[np.argmax(pred, axis=1)[0]]

print(f"Prediction: {pred_name}")
print(f"Actual class: {os.path.basename(os.path.dirname(img_path))}")
```

```python
from lime import lime_image

explainer = lime_image.LimeImageExplainer()

explanation = explainer.explain_instance(image, model.predict, top_labels=1, hide_color=0, num_samples=1000) # 1000 iterations give satisfactory results

predicted_class_int = np.argmax(pred, axis=1)[0]

temp, mask = explanation.get_image_and_mask(predicted_class_int, positive_only=True, num_features=5, hide_rest=True) # Setting num_features=5 because any numberr higher than that makes it hard to catch features

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(temp)
plt.title(f"Highlighted Regions for {pred_name}")
plt.axis('off')
plt.show()
```
